```python
iterator = iter(vg_dataset)
first_item = next(iterator)
second_item = next(iterator)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(first_item["image"])
axes[0].axis("off")
axes[0].set_title(f"Image ID: {first_item['image_id']}")

axes[1].imshow(second_item["image"])
axes[1].axis("off")
axes[1].set_title(f"Image ID: {second_item['image_id']}")

plt.show()
```

```python
print("First Image ID:", first_item["image_id"])
print("First Image URL:", first_item["url"])
print("First Image Regions:", first_item["regions"][:3])
```

```python
print("Second Image ID:", second_item["image_id"])
print("Second Image URL:", second_item["url"])
print("Second Image Regions:", second_item["regions"][:3])
```
